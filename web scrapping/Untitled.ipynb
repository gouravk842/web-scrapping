```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
html_text=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').text
```

```python
soup=BeautifulSoup(html_text,'lxml')
```

```python
job=soup.find('li',class_ ='clearfix job-bx wht-shd-bx')
```

```python
company_name=job.find('h3',class_='joblist-comp-name').text.replace('\r\n','')
```

```python
company_name.replace(' ','')
```

```python
skill=job.find('span',class_='srp-skills')
```

```python
skill.text.replace(' ','').replace("\n",'').replace('\r','')
```

```python
jobs=soup.find_all('li',class_ ='clearfix job-bx wht-shd-bx')
```

```python
for al in jobs:
    company_name=al.find('h3',class_='joblist-comp-name').text.replace('\r\n','').replace(' ','')
    skill=al.find('span',class_='srp-skills').text.replace(' ','').replace("\n",'').replace('\r','')
    print(company_name,"   ",skill)
```
